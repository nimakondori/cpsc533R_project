# Merge the datasets
### We want one csv files that includes all the data with their proper split (train/val/test)

1. Add a mode column to all datasets

```python
train_data["split"] = "train"
val_data["split"] = "val"
test_data["split"] = "test"
```

2. Merge CSV files

```python
final_lvot_df = train_data.append([val_data, test_data])
```

# Analyze Duplicated Patient IDs
### We want to make sure that all the duplicated patients are in the same split

```python
final_lvot_df.PatientID.value_counts()
```

```python
dup_patient_df = final_lvot_df.loc[final_lvot_df.PatientID.duplicated(keep=False)]
```

```python
dup_patient_df.head(5)
```

```python
# group the dataframe by patient ID and split values and count the occurrences
counts = final_lvot_df.groupby(['PatientID', 'split']).size().reset_index(name='count')
duplicates = counts[counts['count'] > 1]
# group the duplicates by patient ID and count the number of unique splits
split_counts = duplicates.groupby('PatientID')['split'].nunique()
# check if the maximum number of unique splits for any patient ID is greater than 1
if split_counts.max() > 1:
    print('Not all duplicated patient IDs are in the same split')
else:
    print('All duplicated patient IDs are in the same split')
```
